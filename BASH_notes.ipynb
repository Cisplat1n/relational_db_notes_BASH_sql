#### Terminal vs Command line:

The 'command line' is a basic text input interface which allows an end user to enter a series of different commands. The command can then be submitted with the Enter/Return key to perform the command. The command line is normally encountered inside of a terminal window. 

A 'terminal' is NOT a command line. It is a special application type that offers a command line interface to perform system-level commands beyond the basic read/write operations. 

Finally, another alternative is the 'shell'. This is a wrapped that contians the command line and interprets inputs commands which it then returns results to.

#### Simple BASH commands:

There are several basic commands for BASH:

'pwd' => print working directory. This will print to the terminal the current folder layer you are in. e.g. /home/lukes/.local/share/jupyter/kernels/bash, if I was in the folders above and forgot, I could use pwd to visually show myself.


'cd' => change directory. You can specify a directory you wish to move to with / and if you wish to go up ONE level in a directory then you can use (..).


'ls' => list. This lists the contents of the current directory. The contents includes every file of every type. You can use additonal flags to gather more information, -a to show hidden files, -l to show file permissons. It is also possible to use cat and less to view file contents.





It is also possible to make files and folders, we can use the following:

'mkdir' => make directory. This makes a folder in whatever directory you run the command from.

'touch' => touch (make file). You can create file within whatever directory you like. e.g. touch notes.txt (Creates a file called notes in .txt format)

'mv' => move. This allows for files to be moved or to have their names be changed. 

There are some more important to be able to remove certain directories or files. We can do that with the following:

'rm' => remove. Can be used to delete files from a directory. If appended with -r, can remove entire directories. -f is important to force the removal of certain protected files.

'cp' => copy. This allows for files to be copied from one location to another without deleting the original file in the original location (unlike mv). To copy an entire directory you need to pass -r flag. 

Finally, it is possible use the echo command (a rough equivalent to that of print()). echo takes a string argument, wrapped in quotes and then can print it to the terminal. It is possible to chain the echo command to control symbols to get different effects. 

E.g. 

echo "Naomi was here" > readme.md (This allows me to overwrite the contents with that string)

```bash
echo "Naomi was here" > test.md 
```

I can also use echo "Naomi was here" >> readme.md (This allows me to append the contents with that string)

```bash
echo "Naomi was here" >> test.md
```
